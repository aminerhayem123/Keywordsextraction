# import lib

```python
from glob import glob
import numpy as np
import pandas as pd
import nltk
from gensim import corpora
import string
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

# import files

```python
files = glob('Rea*')
files.sort()
files
```

# open data and lower case it

```python
df= pd.concat((pd.read_csv(file,names=['text'],encoding='latin-1',sep='\n') for file in files), ignore_index=True)
df['text']=df['text'].str.lower()
df
```

# clean data from caracteres spéciaux et nombres

```python
df['text']=df['text'].str.replace(r"[\"\'\|\?\=\.\<\>\@\#\*\,\;\:\-\[\]\(\)\«\±\%\/\»\\]", '')
df['text']=df['text'].str.replace("\d+", '')
df['text']=df['text'].str.replace("\x85", '')
df['text']=df['text'].str.replace("\x96", '')
df['text']=df['text'].str.replace("\xa0", '')
df['text']=df['text'].str.replace("\x92", '')
df['text']=df['text'].str.replace("\x95", '')
df['text']=df['text'].str.replace("\x9c", '')
df
```

# delete null Values

```python
df_clean = df[df['text'].notnull()]
df_clean
```

# Stop word est Lemmatizaion Task

```python
stop_words = set(stopwords.words('french')+ stopwords.words('english')+['f']) 
lemmatizer = WordNetLemmatizer()
def stop_wrds_lemma_convert(sentence):
    tokens = [w for w in sentence.split() if not w in stop_words] #stopwords removal
    newString=''
    for i in tokens:                                                 
        newString=newString + lemmatizer.lemmatize(i)+' '    #converting words to lemma                               
    return newString.strip()

df_clean['text']= df_clean['text'].apply(lambda x: stop_wrds_lemma_convert(x))
df_clean
```

# Liste des phrase

```python
List_phrase = []
for i in df_clean.text:
    List_phrase.append(i)
List_phrase
```

# List des Mot

```python
List_Mot = []
for i in df_clean.text:
    for j in i.split():
        List_Mot.append(j)
List_Mot
```

# freq words, et lemmatization sur 5000 rows
# data is so big

```python
lst = List_Mot[0:5000]
lemmatizer = WordNetLemmatizer()
data = pd.DataFrame(columns=['Mot','Fréquence','lemmatizer'])
for d in lst:
    data.loc[len(data.index)] = [d, lst.count(d),lemmatizer.lemmatize(d)]
data
```
